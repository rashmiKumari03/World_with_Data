# Titanic Dataset Description ðŸš¢

The Titanic dataset is a well-known dataset in the field of machine learning and data analysis. It is often used for educational purposes and serves as a beginner-friendly introduction to data exploration, cleaning, and predictive modeling. The dataset contains information about passengers aboard the Titanic, including whether they survived or not.

## Features

1. **PassengerId:** A unique identifier assigned to each passenger.
2. **Survived:** This is the target variable. It indicates whether a passenger survived (1) or did not survive (0).
3. **Pclass (Ticket Class):** The ticket class reflects the socio-economic status of the passenger. It is categorized as 1st class, 2nd class, or 3rd class.
4. **Name:** The name of the passenger.
5. **Sex:** The gender of the passenger (male or female).
6. **Age:** The age of the passenger. This feature may contain missing values.
7. **SibSp (Siblings/Spouses Aboard):** The number of siblings or spouses a passenger had aboard the Titanic.
8. **Parch (Parents/Children Aboard):** The number of parents or children a passenger had aboard the Titanic.
9. **Ticket:** The ticket number.
10. **Fare:** The amount of money the passenger paid for the ticket.
11. **Cabin:** The cabin number where the passenger stayed. This feature may contain missing values.
12. **Embarked:** The port where the passenger boarded the ship. It can take values 'C' (Cherbourg), 'Q' (Queenstown), or 'S' (Southampton).


**GOAL:**
The goal in analyzing this dataset is often to predict whether a passenger survived based on other features. Exploratory Data Analysis (EDA), feature engineering, and machine learning models are commonly applied to gain insights and build predictive models on this dataset.

````python

from scipy.stats import chi2_contingency
columns = categorical_feature

# Perform chi-squared tests for each pair of columns
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        contingency_table = pd.crosstab(data[columns[i]], data[columns[j]])
        chi2, p, _, _ = chi2_contingency(contingency_table)

        
        # Print results
        print(f"Chi-squared value between {columns[i]} and {columns[j]}: {chi2}")
        print(f"P-value: {p}")
        
        # Interpret the results
        if p < 0.05:
            print("There is a significant association.")
        else:
            print("There is no significant association.")
        print()
        
    print('\n')
````

Let's introduce the hypothesis testing framework and then explain how the chi-square test with the p-value leads to the insights:

### Hypothesis Testing Introduction:

Hypothesis testing is a statistical method used to make inferences about a population based on a sample of data. 
In the context of the chi-squared test for association between categorical variables, the process typically involves the following:

1. **Null Hypothesis (\(H_0\)):** Assumes there is no association between the variables
  
2. **Alternative Hypothesis (\(H_a\)):** Assumes there is a significant association between the variables.

### Insights from Chi-Squared Test with P-value:

1. **"Sex" and "Ticket":**
   - **Hypothesis:** \(H_0\): There is no association between "Sex" and "Ticket."
   - **Result:** Chi-squared value of 5.0 with a p-value of 0.082.
   - **Conclusion:** Fail to reject \(H_0\). No significant association found.

2. **"Sex" and "Cabin":**
   - **Hypothesis:** \(H_0\): There is no association between "Sex" and "Cabin."
   - **Result:** Chi-squared value of 2.92 with a p-value of 0.233.
   - **Conclusion:** Fail to reject \(H_0\). No significant association found.

3. **"Sex" and "Embarked":**
   - **Hypothesis:** \(H_0\): There is no association between "Sex" and "Embarked."
   - **Result:** Chi-squared value of 5.0 with a p-value of 0.082.
   - **Conclusion:** Fail to reject \(H_0\). No significant association found.

4. **"Ticket" and "Cabin":**
   - **Hypothesis:** \(H_0\): There is no association between "Ticket" and "Cabin."
   - **Result:** Chi-squared value of 3.75 with a p-value of 0.441.
   - **Conclusion:** Fail to reject \(H_0\). No significant association found.

5. **"Ticket" and "Embarked":**
   - **Hypothesis:** \(H_0\): There is no association between "Ticket" and "Embarked."
   - **Result:** Chi-squared value of 10.0 with a p-value of 0.0404.
   - **Conclusion:** Reject \(H_0\). Significant association found.

6. **"Cabin" and "Embarked":**
   - **Hypothesis:** \(H_0\): There is no association between "Cabin" and "Embarked."
   - **Result:** Chi-squared value of 3.75 with a p-value of 0.441.
   - **Conclusion:** Fail to reject \(H_0\). No significant association found.

### Summary:

- The chi-squared test, combined with the p-value, allows us to assess whether the observed associations are likely due to chance or if there's a significant relationship between categorical variables.
- When the p-value is low (typically < 0.05), we reject the null hypothesis, suggesting a significant association.
- When the p-value is high, we fail to reject the null hypothesis, indicating no significant association.
  

Conclusion : Ticket and Embarked Has Some Association
