```python
data=pd.DataFrame(data=[[1,np.nan,2],[1,3,4]],
                 columns=['C1','C2','C3'],
                  index=['R1','R2'])
```

```python
data
```

```python
data.isnull()
```

```python
data.isnull().sum()  # There is a Nan value in C2 Columns/Feature
```

```python
data.isnull().sum().sum()  # Total Nan Value is 1 
```

```python
data['C2'].mean()
```

```python
data['C3'].median()
```

```python
# To fill this Nan value we need to use fillna() and this can be done using mean or median while the data is numerical value.

median=data['C3'].median()
data['C3'].fillna(median,inplace=True)
```

```python
data=pd.DataFrame(data=np.arange(0,20).reshape(5,4),
             columns=['Col1','Col2','Col3','Col4'],
             index=['Row1','Row2','Row3','Row4','Row5'])
data
```

```python
# We can use Filtering Also using Conditional Operatiors on the Data( FILTER AND MASKING)

data['Col2']>10   # Filter
```

```python
data[data['Col2']>10]  # Filter and Masking to get the values
```
